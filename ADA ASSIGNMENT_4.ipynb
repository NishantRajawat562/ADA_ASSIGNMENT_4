```python
import time

def is_safe(board, row, col, n):
    for i in range(col):
        if board[row][i] == 1:
            return False

    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i, j = row, col
    while i < n and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1

    return True


def solve_nqueens(board, col, n):
    if col >= n:
        return True

    for i in range(n):
        if is_safe(board, i, col, n):
            board[i][col] = 1

            if solve_nqueens(board, col + 1, n):
                return True

            board[i][col] = 0  # backtrack

    return False


n = 4
board = [[0]*n for _ in range(n)]

if solve_nqueens(board, 0, n):
    print("Solution:")
    for row in board:
        print(row)
else:
    print("No Solution")
```

```python
def bound(profit, weight, capacity, items, index):
    if weight >= capacity:
        return 0

    result = profit
    j = index
    total_weight = weight

    while j < len(items) and total_weight + items[j][0] <= capacity:
        total_weight += items[j][0]
        result += items[j][1]
        j += 1

    if j < len(items):
        result += (capacity - total_weight) * items[j][1] / items[j][0]

    return result


def knapsack_bb(items, capacity):
    items.sort(key=lambda x: x[1]/x[0], reverse=True)
    max_profit = 0

    def solve(index, profit, weight):
        nonlocal max_profit

        if weight <= capacity and profit > max_profit:
            max_profit = profit

        if index >= len(items):
            return

        if bound(profit, weight, capacity, items, index) > max_profit:
            solve(index+1, profit + items[index][1], weight + items[index][0])
            solve(index+1, profit, weight)

    solve(0, 0, 0)
    return max_profit


items = [(10, 60), (20, 100), (30, 120)]
capacity = 50

print("Maximum Profit:", knapsack_bb(items, capacity))
```

```python
def naive_search(text, pattern):
    n, m = len(text), len(pattern)
    comparisons = 0

    for i in range(n - m + 1):
        match = True

        for j in range(m):
            comparisons += 1
            if text[i+j] != pattern[j]:
                match = False
                break

        if match:
            print("Pattern found at index", i)

    return comparisons


text = "ABABDABACDABABCABAB"
pattern = "ABABCABAB"

comp = naive_search(text, pattern)
print("Total Comparisons:", comp)
```

```python
def test_performance():
    sizes = [100, 500, 1000, 2000]
    
    for size in sizes:
        text = "A" * size + "B"
        pattern = "AB"
        
        start = time.time()
        comp = naive_search(text, pattern)
        end = time.time()
        
        print(f"\nSize: {size}")
        print("Comparisons:", comp)
        print("Time:", end - start)

test_performance()
```
